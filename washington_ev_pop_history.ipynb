```python
#BEV Graph
plt.figure(figsize = (10, 5))
plt.plot(df['Date'], df['Battery Electric Vehicle (BEV) Count'])
plt.xticks(fontsize = 6, rotation = 90)
```

```python
#BEV Month on month percent change graph
plt.figure(figsize = (10, 5))
plt.plot(df['Date'], df['BEV Percent Change'])
plt.xticks(fontsize = 6, rotation = 90)
```

```python
#Reading in 48 month car loans data
#https://fred.stlouisfed.org/series/TERMCBAUTO48NS
month_loans_48 = pd.read_csv('datasets/economic_data/48_month_loans.csv')
month_loans_48.head()
```

```python
#Going to fill in missing values with previous loan entry
temp = None
for index, value in enumerate(month_loans_48['TERMCBAUTO48NS']):
    if value != '.':
        temp = value  # Update last_number when a number is found
    else:
        if temp is not None:
            month_loans_48.at[index, 'TERMCBAUTO48NS'] = temp

month_loans_48.head()
```

```python
#Reading in 60 month car loans data
#https://fred.stlouisfed.org/series/RIFLPBCIANM60NM
month_loans_60 = pd.read_csv('datasets/economic_data/60_month_loans.csv')
month_loans_60.tail()
```

```python
#Going to fill in missing values with previous loan entry
temp = None
for index, value in enumerate(month_loans_60['RIFLPBCIANM60NM']):
    if value != '.':
        temp = value  # Update last_number when a number is found
    else:
        if temp is not None:
            month_loans_60.at[index, 'RIFLPBCIANM60NM'] = temp

month_loans_60.tail()
```

```python
#Only going to keep from 2017-01-01 and beyond
month_loans_48['DATE'] = pd.to_datetime(month_loans_48['DATE'])
month_loans_48 = month_loans_48[month_loans_48['DATE'] >= '2017-01-01']
month_loans_48.head()

month_loans_60['DATE'] = pd.to_datetime(month_loans_60['DATE'])
month_loans_60 = month_loans_60[month_loans_60['DATE'] >= '2017-01-01']
month_loans_60.head()
```

```python
month_loans_48 = month_loans_48.reset_index(drop=True)
month_loans_60 = month_loans_60.reset_index(drop=True)
month_loans_48.tail()
```

```python
#Going to delete last row in main df to match length of car loans data
df.drop(80, inplace = True)
```

```python
#Adding both monthly loans to dataframe
df['48_month_loan'] = month_loans_48['TERMCBAUTO48NS']
df['60_month_loan'] = month_loans_60['RIFLPBCIANM60NM']
```
